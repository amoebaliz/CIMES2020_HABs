These two functions produced the same answer. Here's an example of a dataset that has multiple dimensions and it becomes important to specify how you average:

```python
small_data_set = np.empty((360,2,2))
small_data_set[:,0,0]= .5*np.sin((2*np.pi*mns)/12) + mns/100 
small_data_set[:,0,1]= .5*np.sin((2*np.pi*mns)/12) + mns/1000 
small_data_set[:,1,0]= .25*np.sin((2*np.pi*mns)/12) + 2*mns/100 
small_data_set[:,1,1]= .25*np.sin((2*np.pi*mns)/12) + 4 
```

Ok - so what I did was first create an empty array using "np.empty". I then specified the shape of that array: 360 in the time dimension, and 2 x 2 in the spatial dimension. Those dimensions don't have to be time and space - they could be anything but I am pretending here that it is a set of 4 grid cells on a map that I might want to study. 

I then fill it in with some not-so-random data. the information that appears in the "[]" are the indexes or the matrix locations where I want to put the data I made up. the ":" means all of the places or locations in that dimension - I am filling in a time series so I have all of the time points in my fake little data set. then, I specify the lat/lon position. Python indexing starts with "0"; whatever's in the lower left corner of the matrix has the address "[0,0]"

Let's plot them on a time series plot to see how they all look. Again, we're pulling the time series out of each grid cell:

```python
plt.figure(figsize=(20, 4))
plt.plot(mns,small_data_set[:,0,0],'bo')
plt.plot(mns,small_data_set[:,0,1],'rh')
plt.plot(mns,small_data_set[:,1,0],'g-')
plt.plot(mns,small_data_set[:,1,1],'mx')
plt.xlabel('Month Number')
plt.ylabel('Data Value')
```

so - all a little different. Why might that be? Well, depending on the variable (temperature for example), maybe one point is closer to the land or the equator. Maybe it's located near a really strong current. There are lots of mechanisms to explore. Now, what happens if we take the mean?

```python
np.mean(small_data_set)
```

Timeseries Trend and Trend Significance: 

Trends is interesting because it is a way to look at how things change over time (kindof fundamental to studying climate change). For this we're going to load a python package that is better suited for stats called SciPy. You can achieve answers using numpy but the equations get kindof long and it's easy to make a mistake. 

```python
# I am going to nickname the package "stats", which is silly because it has the same number of letters as "scipy"
# You can also choose not to nickname (or alias) the packages - totally up to you
from scipy import stats
```

We're going to use the linear regression tool, "linregress" in scipy to calculate the least-squares linear regression: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

This function gives a lot of information; the parts we care about are 1) the slope (i.e., trend) and 2) the p-value (i.e., whether that slope is significantly greate than zero... can we actually detect the trend through the noise)

```python
# For our random data
rnd_slope, rnd_intercept, r_value, p_value, std_err = stats.linregress(mns, rnd_data)
print(rnd_slope)
print(p_value)
```

Ok, so the slope (probably) very small and p-value is (probably) rather large: In science, we usually say something is significant if the p-value is less than 0.05 (to be extra careful people will use a threshold of 0.01 or even 0.001) ... with a p-value greater than 0.05, the trend we calculated is not significant. 

So, even though we found a small slope/ trend in the data set, it's not significant and we can't be sure whether it's a real signal (i.e., change) or whether it's just random noise. Makes sense since this was a randomly generated data set! 

NOTE: new random data will be created everytime np.random.rand() is called so the slope and p-value will be different too. But, because it's random, the trend will likely be small and the p-value not significant 

Now let's see what happens with the non-random data set, "more_data"

```python
mor_slope, mor_intercept, r_value, p_value, std_err = stats.linregress(mns, more_data)
print(mor_slope)
print(p_value)
```

We still get a small slope but the p-value is SUPER TINY, suggesting that that trend is a real signal in the data. 

We can use the information from the calculations above to plot the linear agressions against the timeseries:

```python
plt.figure(figsize=(20, 4))
plt.plot(mns,rnd_data,'go')
plt.plot(mns,more_data,'bo')
# the linear requession slope and intercept are applied to the x-data, "mns"
rnd_reg = rnd_slope*mns + rnd_intercept
mor_reg = mor_slope*mns + mor_intercept

plt.plot(mns,rnd_reg,'g--')
plt.plot(mns,mor_reg,'b--')
plt.xlabel('Month Number')
plt.ylabel('Data Value')
```

Notice, in the random (green) data - the regression line looks a lot like the mean we plotted before. This makes sense as the trend was determined not to be significantly different from the mean so it makes sense that it would look the same when plotted.

The more_data (blue) regression, however, looks very different from the more_data (blue) mean because it is representing the trend in the data. 

Now back to plotting a larger data set with "small_data_set":

Here we have to do things a bit differently. Before, with taking the mean and standard deviation, we could tell numpy the axis=0 in order to get the time-mean for all timeseries in the 2x2 grid. We can't do that with linregress. What we will do instead is go through (i.e., "itterate over") all 4 of our grid cells and calculate the slopes and p-values one by one for all four timeseries. To do this we will use "for loops":

```python
# First we create some empty 2x2 storage variables that we will fill with the calculated slopes and p-values
slopes = np.empty((2,2))
p_values = np.empty((2,2))

# Now we're going to go over the index values for latitude (0 and 1)
for nlat in range(lats.shape[0]):
    # and for each latitude index we go over the index values for longitude (0 and 1)
    for nlon in range(lons.shape[1]):
        # I am taking the time series, all [:] 360 values at each lat/lon index
        single_timeseries = small_data_set[:,nlat,nlon]
        # now I calculate the linear regression and store the results in the empty matrices I created before the for loops
        # Note: the symbol "_" tells python that I don't care about that particular output value 
        #       the linregress function calculates it anyway but I am telling python not to keep 
        #       it or assign it to a variable
        slopes[nlat,nlon],_,_,p_values[nlat,nlon],_ =  stats.linregress(mns, single_timeseries)
print("slopes or trends = ")        
print(slopes)
print("p-values = ")
print(p_values)
print("slopes or trends = ")        
print(slopes)
print("p-values = ")
print(p_values)
```

Interesting - so small trends but three of them are significant. Remember, that the units for the trend are the data units per month (for example, degrees Celcius per month). Often, trends, even significant trends in the climate system would be pretty small on a month by month basis. It's not until you look at lots of months together that the signal comes out.

To make sure that the color white represents the value of 0, we set the vmin and vmax to be equal in magnitude but opposite in sign (i.e., vmin = -1*vmax)

```python
lats_stip = lats
lons_stip = lons

lats_stip[p_values>0.01]=np.nan
lons_stip[p_values>0.01]=np.nan
print("lats =")
print(lats_stip)
print("lons =")
print(lons_stip)
```

So what about xarray?

Xarray is a package that uses numpy behind the scenes to run the same calculations but in a way that will hopefully make less work for us when using really big data sets. But it's important to understand what's going on under the hood, so to speak
